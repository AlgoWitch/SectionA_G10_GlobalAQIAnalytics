```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/global_air_quality_datacleaning.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df['Country'] = df['Country'].str.strip().str.lower()
```

```python
df['Country'].replace(['-', '?', 'missing', 'unknown'], np.nan, inplace=True)
```

```python
df['City'] = df['City'].str.strip().str.lower()
```

```python
df['City'].replace(['?', '-', 'missing'], np.nan, inplace=True)
```

```python
df['City'].replace({
    'kolktaa': 'kolkata',
    'vanocuver': 'vancouver',
    'mainla': 'manila',
    'madird': 'madrid',
    'marsielle': 'marseille',
    'yokhoama': 'yokohama',
    'sao pualo': 'sao paulo',
    'tamlae': 'tamale',
    'barsa': 'barcelona',
    'ngaoya': 'nagoya',
    'vatseras': 'vasteras',
    'al ian': 'al ain',
    'pzonan': 'poznan',
    'turjillo': 'trujillo',
    'antofagsata': 'antofagasta',
    'oaska': 'osaka',
    'niarobi': 'nairobi',
    'rwaalpindi': 'rawalpindi',
    'kmuasi': 'kumasi',
    'shnaghai': 'shanghai',
    'joahnnesburg': 'johannesburg',
    'brsailia': 'brasilia',
    'amstredam': 'amsterdam',
    'saint peterbsurg': 'saint petersburg',
    'buchraest': 'bucharest',
    'adids ababa': 'addis ababa'
}, inplace=True)
```

```python
df['Country'].unique()
```

```python
cols = [
    'AQI','PM2.5','PM10','Deforestation_Rate_%','Afforestation_Rate_%',
    'Vehicles_Increase_%','Industries_Increase_%','Env_Budget_Million_USD',
    'Population_Density_Per_SqKm','CO2_Emissions_MT',
    'Green_Space_Ratio_%','Avg_Life_Expectancy_Index'
]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# zero / negative values
for col in cols:
    df.loc[df[col] <= 0, col] = np.nan
```

```python
df.loc[(df['Year'] < 2000) | (df['Year'] > 2025), 'Year'] = np.nan
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.describe()
```

```python
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df.loc[(df[col] < lower) | (df[col] > upper), col] = np.nan
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Country', 'City'])
df['Year'] = df['Year'].fillna(df['Year'].median())
```

```python
# COUNTRY-WISE (environment data)
country_cols = [
    'AQI', 'PM2.5', 'PM10',
    'CO2_Emissions_MT',
    'Green_Space_Ratio_%'
]
```

```python
for col in country_cols:
    df[col] = df.groupby('Country')[col].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
# YEAR-WISE (growth / trend data)
year_cols = [
    'Vehicles_Increase_%',
    'Industries_Increase_%',
    'Env_Budget_Million_USD'
]
```

```python
for col in year_cols:
    df[col] = df.groupby('Year')[col].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
# GLOBAL MEDIAN (remaining columns)
remaining_cols = [
    'Deforestation_Rate_%',
    'Afforestation_Rate_%',
    'Population_Density_Per_SqKm',
    'Avg_Life_Expectancy_Index'
]
```

```python
for col in remaining_cols:
    df[col].fillna(df[col].median(), inplace=True)
```
